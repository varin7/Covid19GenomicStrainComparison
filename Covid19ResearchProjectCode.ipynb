```r
#put data as matrix
div_data <- c(0, 0.00224, 0.0029001, 0.00369, 0.00369, 0.003053, 0.003397, 0.00364, 0.001198,
              0.00224, 0, 0.00305, 0.00386, 0.00389, 0.00329, 0.00356, 0.00384, 0.00137,
              0.0029001, 0.00305, 0, 0.002703, 0.00282, 0.00221, 0.00251, 0.00266, 0.002205,
              0.00369, 0.00386, 0.002703, 0, 0.00178, 0.00164, 0.00185, 0.000313, 0.003102,
              0.00369, 0.00389, 0.00282, 0.00178, 0, 0.00161, 0.00195, 0.00171, 0.00311,
              0.003053, 0.00329, 0.00221, 0.00164, 0.00161, 0, 0.001021, 0.00157, 0.00249,
              0.003397, 0.00356, 0.00251, 0.00185, 0.00195, 0.001021, 0, 0.001804, 0.00283,
              0.00364, 0.00384, 0.00266, 0.000313, 0.00171, 0.00157, 0.001804, 0, 0.003066,
              0.001198, 0.00137, 0.002205, 0.003102, 0.00311, 0.00249, 0.00283, 0.003066, 0)
names = c("B.1.1.7", "B.1.617.2", "BA.1", "BA.2", "BA.2.12.1", "BA.5",
          "BQ.1.1", "XBB", "Wuhan")
divmatrix <- matrix(data = div_data, nrow = 9, ncol = 9, byrow = TRUE,
                    dimnames = list(names, names))
```

```r
#use matrix to create figure 1
pheatmap(divmatrix)
```

```r
#Input Sequence variants
divergence2 <- c(0.001198, 0.00137, 0.002205, 0.003102, 0.00311, 0.00249,
                 0.00283, 0.003066)

time2 <- c(448, 532, 714, 798, 854, 896, 1050, 1106)

df2 <- data.frame(divergence2, time2)
lm2 <- lm(divergence2~time2, data = df2)
summary(lm2)
par(mar = c(5, 6, 4, 2) + 0.1)
plot(divergence2~time2, data = df2,
     xlab = "Days between emergence of Wuhan strain and dominant strain takeovers",
     ylab = expression(paste("Divergence: ", frac("# differing sites",
                                                  "total aligned sites"))),
     pch = 17, col = "blue")
abline(lm2)
```

```r
#Input wuhan vs all variant values

divergence1 <- c(0.001198, 0.00224, 0.00305, 0.002703, 0.00178, 0.00161,
                 0.001021, 0.001804)
time1 <- c(448, 84, 182, 84, 56, 42, 154, 56)

df1 <- data.frame(divergence1, time1)
lm1 <- lm(divergence1~time1, data = df1)
summary(lm1)
par(mar = c(5, 6, 4, 2) + 0.1)
plot(divergence1~time1, data = df1,
     xlab = "Days between dominant strain takeover",
     ylab = expression(paste("Divergence: ", frac("# differing sites",
                                                  "total aligned sites"))),
     pch = 17, col = "red")
abline(lm1)
```
